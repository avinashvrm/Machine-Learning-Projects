```python
df = pd.read_excel('Data.xlsx',sheet_name='Train_Data')
df
```

```python
# Highest Runs in numerial values
df['HS'] = df['HS'].apply(lambda x: str(x).replace('*','') if '*' in str(x) else str(x))
df['HS'] = df['HS'].apply(lambda x: float(x))
```

```python
df.head()
# removed "*"
```

```python
# In Cricket Avg Runs = Total Run / (Number of time OUT) = inn - number of time not out.
df[df['Avg']=='-'] = 0
```

```python
df.drop("PLAYER", axis = 1,inplace = True)
```

```python
target = df['2019_Runs']
```

```python
df.drop("2019_Runs", axis = 1,inplace = True)
```

```python
df.head()
```

```python
#Annotated Heatmaps of a Correlation Matrix
df_corr = df.corr()
```

```python
#hide the upper triangle as they are similar
mask = np.zeros_like(df_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)]= True
```

```python
f, ax = plt.subplots(figsize=(11, 15)) 
heatmap = sns.heatmap(df_corr, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap = 'coolwarm',
                      cbar_kws = {'shrink': .4, 
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1, 
                      vmax = 1,
                      annot = True,
                      annot_kws = {"size": 12})
#add the column names as labels
ax.set_yticklabels(df_corr.columns, rotation = 0)
ax.set_xticklabels(df_corr.columns)
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```

```python
plt.subplots(figsize=(6,7)) 
plt.scatter(df['SR'],df['2018_Runs'],color='red',label='runs', s=50)
plt.xlabel("SR")
plt.ylabel("RUNS")
plt.title('RUNS/SR')
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,target,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=70)
model.fit(X_train,y_train.ravel())
model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X_train,y_train)
regr_2.fit(X_train,y_train)
```

```python
regr_1.score(X_test,y_test)
```

```python
regr_2.score(X_test,y_test)
```

```python
knn = KNeighborsRegressor(n_neighbors = 12)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVR
s_model = SVR(kernel='linear', C=1100,)
s_model.fit(X_train,y_train)
s_model.score(X_test,y_test)
```

```python
from sklearn.svm import SVR
ss_model = SVR(kernel='poly', C=1100,)
ss_model.fit(X_train,y_train)
ss_model.score(X_test,y_test)
```

```python
# SVR MODEL PERFORMS THE BEST WITH KERNEL = 'POLY' AND REGULARISATION = 1100. 
```
